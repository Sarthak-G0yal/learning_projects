# Trajactory Visualization Of Planet And Satellite Using Python
## Introduction
The **Jupyter Notebook File** contains an analysis of planetary and satellite positions and trajectories using Python. **Earth** is the main focus, with the satellites being the **Moon and Chandrayaan**. The data is presented in simplified fractional form for clarity and ease of understanding, aiding in the visualization of orbital dynamics.

```python
import numpy as np
import matplotlib.pyplot as plt

class planet():
    def __init__(self,distPlanet,AngularVelocityPlanet):
        self.rp = distPlanet
        self.wp = AngularVelocityPlanet
        self.t = np.arange(0,3.2,0.02)

    def Get_x(self,r,w):
        return r*np.cos(w*self.t)
        
    def Get_y(self,r,w):
        return r*np.sin(w*self.t)    

    def PlotPlanet(self):
        Xp = self.Get_x(self.rp,self.wp)
        Yp = self.Get_y(self.rp,self.wp)
        plt.plot(1,0,'o',color = '#ff8f00')
        plt.plot(Xp,Yp,color = '#12dbe5')
        plt.grid(True)
        plt.show()
        
```

```python
Earth = planet(4.0,2.0)
Earth.PlotPlanet()
```

```python
class moon(planet):
    def __init__(self,distPlanet,AngularVelocityPlanet,distmoon,AngularVelocityMoon):
        planet.__init__(self,distPlanet,AngularVelocityPlanet)
        self.rm = distmoon
        self.wm = AngularVelocityMoon
    
    def PlotMoon(self):
         self.Xm = self.Get_x(self.rp,self.wp) + self.Get_x(self.rm,self.wm)
         self.Ym = self.Get_y(self.rp,self.wp) + self.Get_y(self.rm,self.wm)
         plt.plot(self.Get_x(self.rp,self.wp),self.Get_y(self.rp,self.wp),color = '#12dbe5')
         plt.plot(self.Xm,self.Ym,color = '#808080')
         plt.plot(1,0,'o',color = '#ff8f00')
         plt.grid(True)
         plt.show()
        
```

```python
EarthMoon = moon(4.0,2.0,1.0,14.0)
EarthMoon.PlotMoon()
```

```python
class satellite(moon):
    def __init__(self,distPlanet,AngularVelocityPlanet,distmoon,AngularVelocityMoon,satdist,AngularVelocitySat):
        moon.__init__(self,distPlanet,AngularVelocityPlanet,distmoon,AngularVelocityMoon)
        self.rs = satdist
        self.ws = AngularVelocitySat

    def PlotSat(self):
         self.Xm = self.Get_x(self.rp,self.wp) + self.Get_x(self.rm,self.wm)
         self.Ym = self.Get_y(self.rp,self.wp) + self.Get_y(self.rm,self.wm)
         self.Xs = self.Get_x(self.rp,self.wp) + self.Get_x(self.rm,self.wm) + self.Get_x(self.rs,self.ws)
         self.Ys = self.Get_y(self.rp,self.wp) + self.Get_y(self.rm,self.wm) + self.Get_y(self.rs,self.ws)
         plt.plot(self.Get_x(self.rp,self.wp),self.Get_y(self.rp,self.wp),color = '#12dbe5')
         plt.plot(self.Xs,self.Ys,color = '#000000')
         plt.plot(self.Xm,self.Ym,color = '#808080')
         plt.plot(1,0,'o',color = '#ff8f00')
         plt.legend(["Earth","Chandrayan","Moon","Sun"],loc = 'upper right')
         plt.grid(True)
         plt.show()          
```

```python
chandrayan = satellite(4.0,2.0,1.0,14.0,.04,100)
chandrayan.PlotSat()
```
